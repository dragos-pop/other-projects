```python
# gravitational acceleration = accelerometer - linear accelerometer
df["grav_acc_avg_X"] = df["acc_avg_X (m/s^2)"] - df["lin_acc_avg_X (m/s^2)"]
df["grav_acc_avg_Y"] = df["acc_avg_Y (m/s^2)"] - df["lin_acc_avg_Y (m/s^2)"]
df["grav_acc_avg_Z"] = df["acc_avg_Z (m/s^2)"] - df["lin_acc_avg_Z (m/s^2)"]

df["grav_acc_min_X"] = df["acc_min_X (m/s^2)"] - df["lin_acc_min_X (m/s^2)"]
df["grav_acc_min_Y"] = df["acc_min_Y (m/s^2)"] - df["lin_acc_min_Y (m/s^2)"]
df["grav_acc_min_Z"] = df["acc_min_Z (m/s^2)"] - df["lin_acc_min_Z (m/s^2)"]

df["grav_acc_max_X"] = df["acc_max_X (m/s^2)"] - df["lin_acc_max_X (m/s^2)"]
df["grav_acc_max_Y"] = df["acc_max_Y (m/s^2)"] - df["lin_acc_max_Y (m/s^2)"]
df["grav_acc_max_Z"] = df["acc_max_Z (m/s^2)"] - df["lin_acc_max_Z (m/s^2)"]
```

```python
df
```

```python
# sliding windows (previous change)

for x in df.columns:
    df['{}_T-1'.format(x)] = df[x].shift(1)
```

```python
df
```

```python
df.columns[:30]
```

```python
# moving averages (of the average aggregations only)
df['SMA10_acc_avg_X'] = df['acc_avg_X (m/s^2)'].rolling(10).mean()
df['SMA10_acc_avg_Y'] = df['acc_avg_Y (m/s^2)'].rolling(10).mean()
df['SMA10_acc_avg_Z'] = df['acc_avg_Z (m/s^2)'].rolling(10).mean()

df['SMA10_gyr_avg_X'] = df['gyr_avg_X (rad/s)'].rolling(10).mean()
df['SMA10_gyr_avg_Y'] = df['gyr_avg_Y (rad/s)'].rolling(10).mean()
df['SMA10_gyr_avg_Z'] = df['gyr_avg_Z (rad/s)'].rolling(10).mean()

df['SMA10_lin_acc_avg_X'] = df['lin_acc_avg_X (m/s^2)'].rolling(10).mean()
df['SMA10_lin_acc_avg_Y'] = df['lin_acc_avg_Y (m/s^2)'].rolling(10).mean()
df['SMA10_lin_acc_avg_Z'] = df['lin_acc_avg_Z (m/s^2)'].rolling(10).mean()

df['SMA10_mag_avg_X'] = df['mag_avg_X (µT)'].rolling(10).mean()
df['SMA10_mag_avg_Y'] = df['mag_avg_Y (µT)'].rolling(10).mean()
df['SMA10_mag_avg_Z'] = df['mag_avg_Z (µT)'].rolling(10).mean()
```

```python
df = df.dropna()
```
